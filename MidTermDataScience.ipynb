####Percentage of products with rating less than 3.5:

```python
rating_below_3_5 = sneakers[sneakers['Rating'] < 3.5]
percentage_rating_below_3_5 = (len(rating_below_3_5) / len(sneakers)) * 100
```

```python
print(f"Percentage of products with listing prices less than 15000: {percentage_listing_price_below_15000:.2f}%")
print(f"Percentage of products with sale price less than 15000: {percentage_sale_price_below_15000:.2f}%")
print(f"Percentage of products with at least 50% discount: {percentage_discount_at_least_50:.2f}%")
print(f"Percentage of products with rating less than 3.5: {percentage_rating_below_3_5:.2f}%")
print(f"Percentage of products with review numbers less than 100: {percentage_reviews_below_100:.2f}%")
print(f"The 3 highest percentages of the discount graph: {percentage_top_3_discounts[0]:.2f}%, {percentage_top_3_discounts[1]:.2f}%, {percentage_top_3_discounts[2]:.2f}%")
```

####The two highly positively correlated variables and The two highly negatively correlated variables

```python
numeric_columns = sneakers.select_dtypes(include=[pd.np.number]).columns
correlation_matrix = sneakers[numeric_columns].corr()

# Create a masked version of the correlation matrix to exclude diagonal elements
masked_corr_matrix = correlation_matrix.copy()
np.fill_diagonal(masked_corr_matrix.values, np.nan)

# Find the most positively correlated variables
max_corr = masked_corr_matrix.stack().idxmax()
max_corr_value = masked_corr_matrix.stack().max()

# Find the most negatively correlated variables
min_corr = masked_corr_matrix.stack().idxmin()
min_corr_value = masked_corr_matrix.stack().min()

print(f"The two most positively correlated variables are '{max_corr[0]}' and '{max_corr[1]}' with a correlation of {max_corr_value:.2f}.")
print(f"The two most negatively correlated variables are '{min_corr[0]}' and '{min_corr[1]}' with a correlation of {min_corr_value:.2f}.")
```

#####Which company gives more discount: Adidas or Nike?

```python
adidas_discount = sneakers[sneakers['Brand'].str.contains('Adidas', case=False)]['Discount'].mean()
nike_discount = sneakers[sneakers['Brand'].str.contains('Nike', case=False)]['Discount'].mean()

if adidas_discount > nike_discount:
    print(f"Adidas gives more discount with an average discount of {adidas_discount:.2f}% compared to Nike's {nike_discount:.2f}%.")
elif nike_discount > adidas_discount:
    print(f"Nike gives more discount with an average discount of {nike_discount:.2f}% compared to Adidas' {adidas_discount:.2f}%.")
else:
    print(f"Both Adidas and Nike give the same average discount of {adidas_discount:.2f}%.")
```
